```python
import numpy as np
from citipy import citipy

from config import weather_api_key 
import requests
import pandas as pd

print(weather_api_key)
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
coordinates
```

```python
# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count.
len(cities)
#print(cities)
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?&units=Imperial&APPID=" + weather_api_key
```

```python
city_info = []

#city_testing = ['livermore']
#for city in cities:


# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")


for city in cities:
    query_url = url + "&q=" + city.replace(" ","+")
    
    try:
        response = requests.get(query_url).json()
        #print(response['name'])
        city_name = response['name']
        country = response['sys']['country']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp= response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response["clouds"]["all"]
        windspeed = response["wind"]["speed"]
        current_descriptions = response['weather'][0]['description']
        city_info.append(
                    {
                       "City": city_name,
                       "Country": country, 
                       "Lat": lat,
                       "Lng": lng,
                       "Max Temp": max_temp,
                       "Humidity": humidity,
                       "Cloudiness": cloudiness,
                       "Wind Speed": windspeed,
                       "Current Descriptions": current_descriptions 
                    }
        )
    except:
        print("City not found. Skipping...")
        pass

    
# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")

print(len(city_info))
```

```python
 
city_weather_info_df = pd.DataFrame(city_info)
city_weather_info_df.head(10)
```

```python
output_file = "WeatherPy_Database.csv"
city_weather_info_df.to_csv(output_file)
```
